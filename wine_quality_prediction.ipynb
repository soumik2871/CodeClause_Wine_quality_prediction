```python
def diagnostic_plots(df, variable,target):
    plt.figure(figsize=(30, 6))

    # histogram
    plt.subplot(1, 4, 1)
    sns.histplot(df[variable], bins=30,color='rebeccapurple')
    plt.title('Histogram')

    # scatterplot
    plt.subplot(1, 4, 2)
    plt.scatter(df[variable],df[target],color='goldenrod')
    plt.title('Scatterplot')
    
    # boxplot
    plt.subplot(1, 4, 3)
    sns.boxplot(y=df[variable],color='green')
    plt.title('Boxplot')
    
    # barplot
    plt.subplot(1, 4, 4)
    sns.barplot(x = target, y = variable, data = df)   
    plt.title('Barplot')
    
    plt.show()
```

```python
def detect_outliers(df,features):
    outlier_indices = []
    for c in features:
        Q1 = np.percentile(df[c],25)
        Q3 = np.percentile(df[c],75)
        IQR = Q3 - Q1
        outlier_step = IQR * 1.5
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2)
    
    return multiple_outliers
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('quality', axis=1),df['quality'],test_size=0.3,random_state=0)
X_train.shape, X_test.shape
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
dct = DecisionTreeClassifier(random_state = 42)
svc = SVC(random_state = 42)
rf = RandomForestClassifier(random_state = 42)
logreg = LogisticRegression(random_state = 42)
knn = KNeighborsClassifier()
sgd = SGDClassifier()
```

```python
classifiers = [("knn" , knn),("rf" , rf),("logreg" , logreg),("svc", svc),("sgd",sgd),("dct",dct)]
```

```python
from sklearn.metrics import accuracy_score
```

```python
for clf_name, clf in classifiers:
    # Fit clf to the training set
    clf.fit(X_train, y_train)
    
    # Predict y_pred
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    # Evaluate clf's accuracy on the test set
    print('{:s} score : {:.3f}'.format(clf_name, acc))
```

```python
param_grid ={'n_estimators' : [50,100,200],'max_depth': (1,5,10),'min_samples_leaf': (1,5,10)}
```

```python
gridsearch = GridSearchCV(rf, param_grid=param_grid, scoring='accuracy', cv=5,n_jobs=6)
```
